```python
# 使用replace 方法
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv')
```

```python
df
```

```python
# 单值替换
```

```python
newdf = df.replace(-99999,value=np.NaN)
```

```python
newdf
```

```python
df
```

```python
df.iloc[4,2]=-88888
```

```python
df
```

```python
newdf = df.replace(to_replace=[-99999,-88888],value=np.NaN)
```

```python
newdf
```

```python
# 替换专门列
```

```python
newdf= df.replace(
{
    'temperature':-99999,
    'windspeed':[ -99999,-88888],
    'event':'0'
},np.nan
)
```

```python
newdf
```

```python
type(df.event[0])
```

```python
# 使用map表进行数据替换
```

```python
df
```

```python
newdf = df.replace({
    
    -99999:np.nan,
   '0': 'No Event',
    -88888: 10
    
})
```

```python
newdf
```

```python
# 正则替换法
```

```python
df
```

```python
newdf = df.replace({
    'temperature':'-[0-9]'
    
},'',regex=True)
```

```python
newdf
```

```python
newdf.iloc[3,1]='abcd'
```

```python
newdf2 = newdf.replace({
    'temperature':'[A-Za-z]'
    
},'',regex=True)
```

```python
newdf2
```
