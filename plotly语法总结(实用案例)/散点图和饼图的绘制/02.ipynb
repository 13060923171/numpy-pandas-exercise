```python
# 1 产生数据
# 2 把数据放入go对象
# 3 创建一个data变量存放go对象
# 4 （可选） data变量是一个数组列表，所以可以存放超过一个以上的go对象
# 5 利用iplot进行绘制
```

```python
groups = ['饮食','账单','娱乐','其他']
```

```python
amount =[1000,500,1500,300]
```

```python
colors = ['#d32c58','#f9b1ee','#b7f9b1','#b1f5f9']
```

```python
trace = go.Pie( labels= groups , values= amount 
              ,hoverinfo='label+percent' ,textinfo = 'value' , textfont= dict(size=25), marker=dict(colors=colors,line=dict(color='#000000' ,width=3)))
data = [trace]
```

```python
iplot(data)
```
